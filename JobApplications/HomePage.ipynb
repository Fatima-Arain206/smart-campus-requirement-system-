```python


def   apply_applicant(name:str , surname:str, age:int,gender:str, roll_no:str , department:str , study_year: int, University_name:str , cgpa:float, city:str ,expe:bool)->None:
    """
    apply for a job
     this annotation function which return nothing
    """
    try:
        cgpa_convert: int = int(cgpa*100)
        ## variable type annotation
        sql_q="INSERT INTO applicant_data(Name,Surname,Age ,Gender,Roll_no,Department,Study_Year,University,CGPA,City,Experience) " \
        "values(%s, %s ,%s, %s ,%s,%s, %s, %s, %s, %s,%s   )"
        user_data=(name,surname,age,gender,roll_no,department,study_year,University_name,cgpa_convert,city,expe)
        my_cursor.execute(sql_q,user_data)
        mydb.commit()
    except mysql.connector.Error as err:

        if err.errno==1062:
            print(f"{roll_no} is already in database kindly check again")
    except Exception as e:
        print(f"{e}  error try again")
    else:

        print(f"your data is  submited ")

# apply_applicant("f")
```
